# Dicts

Other resources: [kx docs](https://code.kx.com/q/basics/dictsandtables/), [oK docs](https://github.com/JohnEarnest/ok/blob/gh-pages/docs/Manual.md), [nsl: Lists are Maps](http://nsl.com/papers/m.txt)

k has a beautiful dict implementation, which should be the envy of any APLer. A k dict is a key-value lookup table that can map anything to anything. To construct a dict in k, we have several options, the easiest is to join a vector of keys to a vector of values with dyadic `!`: [[try it](https://ngn.bitbucket.io/k/#eJzLTq0stkpIyk9KSExJzE0oLk1NKMosSeQqS8wptjIyVjA1UTA0UrAw4soGKlQEiSrgAvoKyfl5xSVFpcklCokKKZnJJQA3sRpl)]

```ngnk
keys:`bob`adam`sue`rita
vals:23 54 12 82
keys!vals                         / construct a dict
```

We can list the keys using `!` and the values with `.`: [[try it](https://ngn.bitbucket.io/k/#eJxLscpOrSxWLEvMKeZSTFEgAPQV0lNLFEA6uPSIVAw0uTS1GABKmRUH)]

```ngnk
d:keys!vals
!d                                / get keys
.d                                / get values
```

and indexing a dict follows the normal k conventions: [[try it](https://ngn.bitbucket.io/k/#eJxLiU4oLk2N5UpxSCjKLEnkSgFTIDEAf+EJsw==)]

```ngnk
d[`sue]
d@`rita
d`rita`sue
```

Updating a dict, unlike updating a vector, has upsert semantics: [[try it](https://ngn.bitbucket.io/k/#eJwFwUsKgDAMBcC9p3g3ED+rXkWEBJNCkVbJBzy+M3KQpxILd1JrF1XjcZ9lwYoNO4AZvX2Rpngq8hUOdfAQtOFq4ZP8KQAXRw==)]

```ngnk
d[`sue`adam`eric`frank]:1 2 3 4   / mixture of updates and inserts
d
```

Unlike dicts found in many other languages, a dict in k is guaranteed to retain its insertion order. 
This is a consequence of its implementation: adding an element to a dict appends the key to the key vector,
and the value to the value vector.

This means that we can take and drop items from the beginning and end, for example, taking the last two k-v pairs: [[try it](https://ngn.bitbucket.io/k/#eJwFwUEOgCAMBMCvrPFsjMWDz9kiJRICJBX/78wmK+OI1KSN72f0MpXm5WZ27XWRAMGBSxBwAjumVkP20TAfg/X0A/CiFVU=)]

```ngnk
-2#`bob`adam`sue`rita`eric`frank!23 2 1 82 3 4  / take from the end
```

As we stated above, both keys and values can be any valid k-type. For example, we frequently need values to be vectors: [[try it](https://ngn.bitbucket.io/k/#eJwrTsxJLbZK8M3PSwgpTU0IT01JCMkoTXArylTUMFQwUbBQsASSptbGCmbW5goGQGhpbQQUBdJAaK5gpmBubQAkjTS5ikFGAQCnNRR9)]

```ngnk
sales:`Mon`Tue`Wed`Thu`Fri!(1 4 8 9 4 5;3 6;7 0 0 9;2 8 0 9 9 7 6 7;0 6 2)
sales
```

In k, a dictionary is treated like any other vector when it comes to function application. Eaching a monad over a dict applies to the values, but leaves the keys, just like eaching a monad over a vector applies to a vector's _elements_, but leaves its indexes. This can be extraordinarily useful. If we say that the above `sales` dict holds product ids sold each day, how many sales did we make per day? [[try it](https://ngn.bitbucket.io/k/#eJxTVi9OzEktVsAD9BUSCwpyKhVy8/MSUzKTFZQVSvIVUhOTMxTKEnNKU3UUilJLEjPzMvPSFbJTK4sBEkgWcA==)]

```ngnk
#'sales                            / apply monadic # to each value, retaining keys
```

If we internalise the concept that a dict is just like a vector, but with a different domain, a lot of clever things follow. For example, we know that in k we can find the indexes of 1s in a boolean vector with `&`: [[try it](https://ngn.bitbucket.io/k/#eJxTM1QwAEJDBRhtAAAbFgLb)]

```ngnk
&1 0 0 1 1 0 0 1 0
```

So what will happen if we apply `&` to a dict with a boolean range? [[try it](https://ngn.bitbucket.io/k/#eJw1w0sKgCAUBdCt3CZNsx29az1TtAR/0e6DoANnlhIaxWYrtatw5ymu8IqTgcH6/S04tIEpIepTcfuweXgOBTGYuiI7rC8r3xlO)]

```ngnk
&`rita`bob`sue`adam`frank!0 0 1 0 1         / get all keys which have a value of 1
```

Similarly, we can use `?` -- which for vectors finds the index of the first occurrence of its right argument in its left -- to do a reverse lookup: find the key for a given value: [[try it](https://ngn.bitbucket.io/k/#eJzTSEjKT0pITEnMTSguTU0oyixJVDQyVjA1UTA0UrAw0rQHUjCgr5CTn59dWqCQnVqpkFSpUJaYU5qqQCQAANZfF/0=)]

```ngnk
(`bob`adam`sue`rita!23 54 12 82)?12         / lookup key by value                                        
```

If there were several keys holding the same value, we'd only get the _first_ such key. This follows from the behaviour of `?` for vectors, and the fact that dicts retain the insertion ordering. If we wanted ALL matching keys, we could do [[try it](https://ngn.bitbucket.io/k/#eJxTM7VNSMpPSkhMScxNKC5NTSjKLElUNFUwVDBVMFZABfoKjjk5CtmplcUKGYllmXnpCmWJOaWpCqYArAITfg==)]

```ngnk
&5=`bob`adam`sue`rita!5 1 5 3               / All keys having value 5
```

What happens if we reduce a dict? Again, thinking of a dict as a vector with a range and domain, let's find the max of a dict: [[try it](https://ngn.bitbucket.io/k/#eJwNwcENgCAQBMBW1goI+MBybhGIFwUTBOPD4nXmNdK0U8IZ5BpJGFkkN9Z9sg4ei8fsYC1+BllrRN8SCh/cPEaCVhBR1/4BjEkXgg==)]

```ngnk
|/`rita`bob`sue`adam`frank!12 7 87 32 11    / find the max value in a dict
```

In other words, reductions (and scans) apply to the _values_ of a dict.

It should come as no surprise that any function application that makes sense on a vector can also be applied to a dict. Keys odd values? Sure (and thus demonstrating three different meanings of `!`): [[try it](https://ngn.bitbucket.io/k/#eJxTM1JMSExISkhOSElITUhLSE/IUFS0UMAK9BWyUyuLFfLTFPJTUhTKEnNKU4sBFrkQ3w==)]

```ngnk
&2!`a`b`c`d`e`f`g`h!!8                      / keys of odd values
```

We could apply `filter` directly to the dict, too, which operates on the values: [[try it](https://ngn.bitbucket.io/k/#eJzTMFLUVE5ITEhKSE5ISUhNSEtIT8hQVLRQwAL0FdIyc0pSixTKEnNKU4sBw9QPaw==)]

```ngnk
(2!)#`a`b`c`d`e`f`g`h!!8                    / filter values
```

The `#` is [special-cased](https://code.kx.com/q/ref/take/#dictionary) for dicts if the left argument is a vector of symbols (or a string in oK and ngn/k), where it becomes `take`: [[try it](https://ngn.bitbucket.io/k/#eJxLSEpIVk5ITABSCSkJqQlpCekJGYqKFgpYgL5CSWJ2qkJJRqpCQpJCYl6KQkKyQkFiZlExALP0E3M=)]

```ngnk
`b`c#`a`b`c`d`e`f`g`h!!8                    / take the `b and `c pairs
```

If you `take` keys that don't exist in the dict, you get a `0N` place-holder: [[try it](https://ngn.bitbucket.io/k/#eJxLSEoITEhWTkhMSEpITkhJSE1IS0hPyFBUtFDAAPoKJYnZqQp5+Xm6qRWZxSWpeSUK2amVCgmBANJfFHI=)]

```ngnk
`b`Q`c#`a`b`c`d`e`f`g`h!!8                  / take non-existent key `Q
```

and, as a final spin on `#`, we can `replicate` (ngn/k-specific): [[try it](https://ngn.bitbucket.io/k/#eJyrNlAwBEIDMASyapUTEhOSEpITUhJSE9IS0hMyFBUtFCBAX6EgMzlboSAxs6hYIalSISk/P0chN7EAAGISE1g=)]

```ngnk
{0 1 1 0 0 0 1 1}#`a`b`c`d`e`f`g`h!!8       / pick pairs by bool map
```

Monadic `=` becomes `group`. If you know APL, this is similar to [key](https://xpqz.github.io/learnapl/key.html) (`⌸`). It takes a vector and creates a mapping from each unique element in order of occurrence to a vector of its indexes of occurrence: [[try it](https://ngn.bitbucket.io/k/#eJyzVUpUSE4sUSgG4vw8hUSF3MQSJQBF/QaD)]

```ngnk
="a cat sat on a mat"
```

`Group` can be applied to any vector, for example [[try it](https://ngn.bitbucket.io/k/#eJyzNVQwUjAEQmMFCwVLMATxLAAzTQQT)]

```ngnk
=1 2 1 1 3 8 9 9 9 1 1 8
```

If we instead wanted frequency counts, we can use the `each-length` approach we used before, [[try it](https://ngn.bitbucket.io/k/#eJxTVrc1VDBSMARCYwULBUswBPEsFJCBvkJOal56SUaxQn6aQnpRfmlBMQAx5gz2)]

```ngnk
#'=1 2 1 1 3 8 9 9 9 1 1 8             / lengths of groups
```

When using `group` like this, occasionally there is the need to ensure that all keys from some set are present, defaulting to some value should they not be present in the data. For example, if we're calculating letter frequencies in a paragraph of text, we might want to make sure that any letters that aren't present in the text are still entered into the table, but with a value of 0. Here's one way we can achieve that using the `take` form of `#` and a "null-fill": [[try it](https://ngn.bitbucket.io/k/#eJx1wUEKgCAUBcB9pzANWkQULVoEXSV6yksDKfkKdfxO0AysW3bXaOiunmb1Z1CeFwWFKt4PpXfIVIgpwLJU4wbrjGlXHSg8MxIEXpDCfRS+JesPSyMeHg==)] [[try it](https://ngn.bitbucket.io/k/#)]

```ngnk
abc:`c$"a"+!26                         / generate lower-case alphabet
0^abc##'="hereisaparagraphoftexts"
```
